```python
def imputation_tech(data_copy):
  from sklearn.impute import SimpleImputer

  print('\n',"Befor imputation")

  print(data_copy.isnull().mean())

  print('\n',"After imputation")

  global data_impute

  data_impute = copy.deepcopy(data_copy)

  impute_strategy = SimpleImputer(strategy='mean')

  data_impute = impute_strategy.fit_transform(data_impute)

  data_impute = pd.DataFrame(data_impute)
  
  data_impute.columns = data_copy.columns

  return data_impute.isnull().sum()
```

```python
imputation_tech(data_copy)
```

```python
def find_outlier(data_impute):
  global q1 ,q3 , iqr
  q1= data_impute.quantile(.25)
  q3= data_impute.quantile(.75)
  iqr= q3-q1

  print(iqr,"\n\n")
  print("less than first quartile")
  print(np.where(data_impute < q1 -iqr *1.5),'\n\n')
  print("greter than 3rd quartile")
  print(np.where(data_impute>q3 + iqr *1.5),'\n\n')
  print("3rd std deviation")
  print(np.where(data_impute> q3+iqr *3))
```

```python
find_outlier(data_impute)
```
